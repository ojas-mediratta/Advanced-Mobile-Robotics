### Code Structure

`gicp.py` contains the source code to perform Generalized-ICP with GTSAM. The main entry point to the program is `gicp()`, which takes in the following parameters:
- *source_points* (np.ndarray): (N, 3) array of points from source point cloud
- *target_points* (np.ndarray): (N, 3) array of points from target point cloud
- *initial_bTa* (gtsam.Pose3 | np.ndarray | None): initial a-to-b SE(3) transformation
- *config* (GICPConfig | None): program parameters

The flow of execution looks like this:
1. Compute covariance matrices for each point in the source and target point clouds
2. Iteratively for specified iterations or until goal reached:
    1. Transform source point cloud to target frame using estimated transformation
    2. Compute nearest-neighbor correspondences
    3. Transform source point covariances to target frame using estimated transformation
    4. Compute information matrices for correspondences
    5. Build and optimize over factor graph to solve for estimated transformation

### Notation

In the program, we reference two frames:
- **source** refers to the scan that is being registered to a later scan. This scan's point cloud resides in the **source** frame (frame **a**).
- **target** refers to the scan that an earlier scan is trying to match. This scan's point cloud resides in the **target** frame (frame **b**).
- **bTa** refers to the SE(3) transformation (**T**) that rotates and translates a point in the **a** frame to the **b** frame.
