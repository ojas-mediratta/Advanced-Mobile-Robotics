## The Exponential Map
We can calculate the exponential map $\exp(\hat{\xi})$ using two common definitions extended to matrices:

1. **The Limit Definition**: 
   Based on the compound interest formula $(1 + \frac{x}{n})^n$, for matrices this becomes:
   $$ \exp(\hat{\xi}) = \lim_{n \to \infty} \left(I + \frac{1}{n}\hat{\xi}\right)^n $$

2. **The Power Series Expansion**: 
   Just as $e^x = \sum \frac{x^n}{n!}$ for scalars, for the matrix $\hat{\xi}$ we have:
   $$ \exp(\hat{\xi}) = I + \hat{\xi} + \frac{1}{2!} \hat{\xi}^2 + \frac{1}{3!} \hat{\xi}^3 + \dots $$

Both converge to the same result, which is the analytic exponential map provided by `gtsam`.

#### Jacobians of Compose
To find the Jacobians of $f(T_1, T_2) = T_1 T_2$, we apply perturbations $\xi_1, \xi_2$ in the tangent space (local coordinates) and look for the resulting perturbation $\zeta$ in the output.

$$ T_{new} = (T_1 \exp(\hat{\xi}_1)) (T_2 \exp(\hat{\xi}_2)) $$

Using the first-order approximation $\exp(\hat{\xi}) \approx I + \hat{\xi}$:
$$ T_{new} \approx T_1 (I + \hat{\xi}_1) T_2 (I + \hat{\xi}_2) $$
$$ T_{new} \approx T_1 T_2 + T_1 \hat{\xi}_1 T_2 + T_1 T_2 \hat{\xi}_2 $$

We want to express this as $T_1 T_2 (I + \hat{\zeta})$. Multiplying the equation by $(T_1 T_2)^{-1} = T_2^{-1} T_1^{-1}$ on the left:

$$ I + \hat{\zeta} \approx I + T_2^{-1} T_1^{-1} T_1 \hat{\xi}_1 T_2 + T_2^{-1} T_1^{-1} T_1 T_2 \hat{\xi}_2 $$
$$ \hat{\zeta} \approx T_2^{-1} \hat{\xi}_1 T_2 + \hat{\xi}_2 $$

The term $T_2^{-1} \hat{\xi}_1 T_2$ corresponds to the **Adjoint map** of $T_2^{-1}$ acting on $\xi_1$. Thus:
$$ \zeta \approx \text{Ad}_{T_2^{-1}} \xi_1 + \xi_2 $$

This gives us the Jacobians:
1. With respect to $T_1$: $\text{Ad}_{T_2^{-1}}$
2. With respect to $T_2$: Identity ($I$)
