GTSAM is used to fit a Fourier basis to the selected points on the $\theta_1(t)$ and $\theta_2(t)$ plots, generating the finer-resolution joint-space trajectory shown in blue. The plots are initialized with five seeded control points at $t=0,1,2,3,4$ (all zero angles) colored red, pink, yellow, green, and blue. Seeded points can only be dragged vertically (their time is fixed). You can still add extra points with mouse clicks and drag those freely. You must have at least 3 points on each joint-angle graph for a fitted trajectory to appear. You can delete points by selecting them and pressing backspace. The fitted trajectory live-updates as you edit points.

Once both joint angles have a fitted trajectory, the path of the foot (computed using closed-form forward kinematics) in Cartesian space is auto-plotted and live-adjusts with the joint-space trajectory. The hip is the black cross (default base height $y_0=3.4$), and the two leg links are shown in black. The task targets now use color and marker coding for accessibility: red circle at $t=0$, pink square at $t=1$, yellow triangle at $t=2$, green diamond at $t=3$, and blue plus-filled marker at $t=4$. Black arrows indicate direction, and intermediate one-second FK markers/ghost poses follow the same color sequence.
